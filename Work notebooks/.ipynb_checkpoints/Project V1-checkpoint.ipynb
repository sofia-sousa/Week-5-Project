## Importing datasets

```python
# Average of the max temperature by year
max_temp = pd.read_excel('/home/inrx/Transferências/Tableau Project/temp-max-media.xlsx', skiprows=7)
max_temp.head()
```

```python
# Max temp by month
max_temp_month = pd.read_excel('/home/inrx/Transferências/Tableau Project/temp-max-mes-quente.xlsx', skiprows=7)
max_temp_month.head()
```

```python
produtividade = pd.read_excel('/home/inrx/Transferências/Tableau Project/Produtividade-aparente-do-trabalho.xlsx',
                             skiprows=6)
produtividade.head()
```

## Cleaning Function

```python
def clean_pordata_temp(df):
    df = df.rename(columns={'Unnamed: 0':'Year'})
    df = df.loc[(df!=0).all(axis=1)]
    df = df.dropna(axis='columns', thresh=2)
    df = df.dropna()
    df = df.reset_index(drop=True)
    return df
```

## Cleaning Data

```python
max_temp_cleaned = clean_pordata_temp(max_temp)
max_temp_cleaned.head()
```

```python
max_temp_month_cleaned = clean_pordata_temp(max_temp_month)
max_temp_month_cleaned.head()
```

```python
seca_init_cleaned = clean_pordata_temp(seca_init)
seca_init_cleaned.head()
```

```python
produtividade_clean = produtividade.dropna(axis='columns', thresh=6)
produtividade_clean = produtividade_clean.dropna()
produtividade_clean = produtividade_clean[:22]
```

## Data Manipulation

### Data manipulation functions

```python
def calc_mean (df, new_column):
    """Calculates de mean of each row, excluding the island territories of Portugal"""
    
    columns = ['Viana do Castelo', 'Bragança', 'Porto', 'Castelo Branco', 'Lisboa', 'Beja', 'Faro']
    df[columns] = df[columns].astype(float)
    df[new_column] = df[columns].mean(axis=1)
    df = df[['Year', new_column]]
    return df
    
```

### Manipulating Data

### Mean of the temperatures and drought

```python
max_temp_final = calc_mean(df=max_temp_cleaned, new_column='avg_max_temp')
max_temp_month = calc_mean(df=max_temp_month_cleaned, new_column='avg_max_temp_month')
drought = calc_mean(df=seca_init_cleaned, new_column='avg_drought')
```

```python
produtividade_final = produtividade_clean.rename(columns={'Anos':'Year'})
produtividade_final['Produtividade aparente do trabalho'] = produtividade_final['Produtividade aparente do trabalho'].astype(float)
```

```python
produtividade_final.head()
```

### Melting Dataframes

```python
max_temp_melted = pd.melt(max_temp_cleaned, id_vars='Year', value_vars=['Viana do Castelo', 'Bragança', 'Porto', 'Castelo Branco', 'Lisboa', 'Beja', 'Faro'], var_name='City',value_name='max_avg_temp')
max_temp_melted.head()
```

```python
max_temp_month_melted = pd.melt(max_temp_month_cleaned, id_vars='Year', value_vars=['Viana do Castelo', 'Bragança', 'Porto', 'Castelo Branco', 'Lisboa', 'Beja', 'Faro'], var_name='City',value_name='max_avg_temp_month')
max_temp_month_melted.head()
```

```python
drought_melted = pd.melt(seca_init_cleaned, id_vars='Year', 
                         value_vars=['Viana do Castelo', 'Bragança', 'Porto', 'Castelo Branco', 'Lisboa', 'Beja', 'Faro'], var_name='City',value_name='max_avg_temp_month')
drought_melted.head()
```

```python
def merge_dataframe (objs, key):
    """Merge DataFrames, given the dfs (objs) and the key (where df will merge)"""
    return reduce(lambda left,right: pd.merge(left,right,on=key), objs)
```

With the merge we will lose some of the information that is present in the temperatures data 

```python
merged_data = merge_dataframe(objs=[drought, max_temp_month, max_temp_final, produtividade_final], key='Year')
merged_data.corr()
```

```python
merged_data.to_csv('merged_dataframe.csv')
```

```python
melted_merged = merge_dataframe(objs=[drought_melted,max_temp_month_melted, max_temp_melted], key=['Year','City'])
```

```python
merged_data2 = merge_dataframe(objs=[melted_merged, produtividade_final], key='Year')
merged_data2.head()
```

```python
merged_data2.to_csv('merged_dataframe2.csv')
```
